```python
# drop columns that are unnecessary and have too many null values
data = data.drop(['Date','Location','Cloud9am','Cloud3pm','Evaporation','Sunshine'],axis=1)
data.head()
```

```python
# drop rows with null values
data = data.dropna(axis=0)
data.shape
```

```python
# some columns like WindGustDir, WindDir9am, WindDir3pm, RainToday, RainTomorrow (target) have string data which cannot be used when modelling
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data['WindGustDir'] = le.fit_transform(data['WindGustDir'])
data['WindDir9am'] = le.fit_transform(data['WindDir9am'])
data['WindDir3pm'] = le.fit_transform(data['WindDir3pm'])
data['RainToday'] = le.fit_transform(data['RainToday'])
data['RainTomorrow'] = le.fit_transform(data['RainTomorrow'])

data.head()
```

```python
# create independent and dependent variables
x = data.drop(['RainTomorrow'],axis=1)
y = data['RainTomorrow']
```

```python
sns.scatterplot(x = 'MaxTemp', y = 'MinTemp', hue = 'RainTomorrow' , palette = 'inferno',data = data)
```

As minimum temperature increases, maximum temperature also increases. It follows a linear relationship.

```python
sns.scatterplot(x = 'Humidity9am', y = 'Temp9am', hue = 'RainTomorrow' , palette = 'inferno',data = data)
```

As humidity increases, the temperature decreases but the chances of rain tomorrow increase. There are more orange dots to the right of the graph.

```python
sns.heatmap(data.corr())
```

The probability of rain tomorrow is highly correlated with Pressure9am, Pressure3pm and Temp3pm. RainToday also affects RainTomorrow

```python
# split the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
predictions = lr.predict(x_test)
print('confusion matrix')
print(confusion_matrix(y_test, predictions))
print()
print('classification report')
print(classification_report(y_test, predictions))
print()
print('accuracy score')
print(accuracy_score(y_test, predictions))
```

#### Has better accuracy than decision tree

# Decision tree

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
predictions = rf.predict(x_test)
print('confusion matrix')
print(confusion_matrix(y_test, predictions))
print()
print('classification report')
print(classification_report(y_test, predictions))
print()
print('accuracy score')
print(accuracy_score(y_test, predictions))
```

#### Has better accuracy than decision tree and logistic regression. Same accuracy as XGboost
